```python
class Node(object):
  def __init__(self,data= None):
    self.data= data
    self.left= None
    self.right= None

#---------------- binary search tree BST ------------------
  def insert_BST(self, new_val):
      if self.data != None:
        if new_val < self.data:
          if self.left is None:
            self.left= Node(new_val)
          else:
            self.left.insert_BST(new_val)
        if new_val > self.data:
          if self.right is None:
            self.right= Node(new_val)
          else:
            self.right.insert_BST(new_val)
        if new_val==self.data:
          return print("this value is already exist in your tree")
      else:
        self.data=Node(new_val)

  def search_BST(self, level, find_val):
      if self.data!=None:
        if find_val == self.data:
          return print(f"{find_val} is found in the level {level} of the tree")
        elif find_val < self.data:
          level +=1
          return self.left.search_BST(level, find_val)
        else:
          level +=1
          return self.right.search_BST(level, find_val)
      return False

#--------- level order (Breadth First Search BFS) ----------
  def insert(self,data): # insert min to max
    if self.data is not None:
      if data < self.data:
        if self.left is None:
          self.left = Node(data)
        else:
          self.left.insert(data)
      elif data > self.data:
        if self.right is None:
          self.right = Node(data)
        else:
          self.right.insert(data)
    else:
      self.data=data 

  def search_Level_Order(self,root, find_val):
    curr=root
    stack=[curr]
    while curr is not None:
        curr=stack.pop(0)
        if curr.data==find_val:
            return True
        if curr.left is not None:
            stack.append(curr.left)
        if curr.right is not None:
            stack.append(curr.right)
    return False

  def Print_Level_Order(self,root):    
    curr=root
    stack=[curr]
    result=[]
    while stack != []:
      result.append(stack[0].data)
      curr= stack.pop(0)

      if curr.left is not None:
        stack.append(curr.left)
      
      if curr.right is not None:
        stack.append(curr.right)
    return result

#---------------------Traversel: Preorder (root-->left-->right)---------------------------
  def Print_Preorder_Recursive(self,root,stack):
    if root is not None:
      stack.append(root.data)
      self.Print_Preorder_Recursive(root.left,stack)
      self.Print_Preorder_Recursive(root.right,stack)
    return stack 

  def Print_Preorder_Iterative(self,root):
    stack=[root]
    result=[]
    while stack != []:
      root= stack.pop()
      result.append(root.data)
      if root.right is not None:
        stack.append(root.right)
      if root.left is not None:
        stack.append(root.left)
    return result

#---------------------Traversel: Inorder (left-->root-->right )---------------------------
  def Print_Inorder_Recursive(self,root,stack): 
    if root is not None:
      if root.left is not None:
        self.Print_Inorder_Recursive(root.left,stack)
      stack.append(root.data)
      self.Print_Inorder_Recursive(root.right,stack)
    return stack
  
  def Print_Inorder_Iterative(self,root):  
    stack = []
    result = []
    while stack != [] or root is not None:
      while root is not None:
        stack.append(root)
        root=root.left
      root=stack.pop()
      result.append(root.data)
      root=root.right
    return result

#---------------------Traversel: Postorder (left-->right-->root)---------------------------
  def Print_Postorder_Recursive(self,root,stack): 
    if root is not None:
      if root.left is not None:
        self.Print_Postorder_Recursive(root.left,stack)
      if root.right is not None:
        self.Print_Postorder_Recursive(root.right,stack)
      stack.append(root.data)
    return stack

  def Print_Postorder_Iterative(self,root):  
    stack1 = [root]
    stack2 = []
    result = []
    while stack1 != []:
      stack2.append(stack1.pop())
      if root.left:
        stack1.append(root.left)
      if root.right:
        stack1.append(root.right)
        root=root.right
      elif stack1 !=[]:
        root=stack1[-1]
    while stack2 != []:
      item=stack2.pop()
      result.append(item.data)
    return result

#---------------------Exercies---------------------------
  def Inorder_successor(self,Node):
    if Node.left is not None:
      self.Inorder_successor(Node.left)
    elif Node.right is not None:
      self.Inorder_successor(Node.right)
    else:
      return Node.data 

  def Delete_Node(self,root,Node_val):
    # Case 1: Node to be deleted is a leaf node. Directly delete the node from the tree. 
    # Case 2: Node to be deleted is an internal node with one child. For this case, delete the node and move its child up to take its place.
    # Case 3: Node to be deleted is an internal node with two children. Copy the contents of the inorder successor of the node to be deleted
    #  and delete the inorder successor. The inorder successor can be found by finding the minimum element in the right subtree of the node. 
    curr=root
    stack=[curr]
    stack_prev=[None]
    while stack != []:
          curr=stack.pop(0)
          prev=stack_prev.pop(0)
          if prev is not None:
            print(curr.data,prev.data)
          if curr.data != Node_val:         
              if curr.left is not None:
                stack.append(curr.left)
                stack_prev.append(curr)  
              if curr.right is not None:
                stack.append(curr.right)
                stack_prev.append(curr)
          else:
              if (curr.left is None) and (curr.right is None):
                  if prev.left==curr:
                    prev.left=None
                  else:
                    prev.right=None
                  curr=None
                  return
              elif (curr.left is not None) and (curr.right is None):
                  curr.data=curr.left.data
                  curr.left=None
                  return 
              elif (curr.left is None) and (curr.right is not None):
                  curr.data=curr.right.data
                  curr.right=None
                  return
              elif (curr.left is not None) and (curr.right is not None):
                  curr.data = self.Inorder_successor(curr.right)
          prev=curr



# Tree0=Node(5)
# Tree0.left=Node(3)
# Tree0.right=Node(8)
# Tree0.left.left=Node(1)
# Tree0.left.right=Node(4)
# Tree0.right.left=Node(7)
# Tree0.right.right=Node(10)

Tree1=Node(5)
Tree1.insert_BST( 3)
Tree1.insert_BST(8)
Tree1.insert_BST(2)
Tree1.insert_BST(1)
Tree1.insert_BST(4)
Tree1.insert_BST(7)
Tree1.insert_BST(10)

print("Print level order: ", Tree1.Print_Level_Order(Tree1))
print("Print preorder using recursive method: ", Tree1.Print_Preorder_Recursive(Tree1,[]))
print("Print preorder using iterative method: ", Tree1.Print_Preorder_Iterative(Tree1))
print("Print Inorder using recursive method: ", Tree1.Print_Inorder_Recursive(Tree1,[]))
print("Print Inorder using iterative method: ", Tree1.Print_Inorder_Iterative(Tree1))
print("Print Postorder using recursive method: ", Tree1.Print_Postorder_Recursive(Tree1,[]))
print("Print Postorder using iterative method: ", Tree1.Print_Postorder_Iterative(Tree1))
Tree1.search_BST(1, 7)
Tree1.search_Level_Order(Tree1, 7)

Tree1.Delete_Node(Tree1,8)
print("Print level order after deleting 10: ", Tree1.Print_Level_Order(Tree1))

Tree1.Delete_Node(Tree1,10)
print("Print level order after deleting 1: ", Tree1.Print_Level_Order(Tree1))

# check whether a given a binary tree is a valid binary search tree (BST) or not.
# find the closest value of a given target value in a given non-empty Binary Search Tree (BST) 
# Sum of all Nodes in a Tree 
# Python Program to Count Leaf Nodes
# Python Program to Count Non-Leaf Nodes 
# Print all Paths from Root to Leaf 
# Print Left View of a Tree 

#           5          
#        /     \
#       3        8
#     /   \    /   \
#    2     4  7    10 
#   /
#  1
```

### Heap

Heap is a special tree structure in which each parent node is less than or equal to its child node. Then it is called a Min Heap.

If each parent node is greater than or equal to its child node then it is called a max heap.

* heapify − This function converts a regular list to a heap. 

* heappush − This function adds an element to the heap without altering the current heap.

* heappop − This function returns the smallest data element from the heap.

* heapreplace − This function replaces the smallest data element with a new value supplied in the function.

```python
import heapq

H = [21,1,45,78,3,5]

# Create the heap
heapq.heapify(H)
print(H)   # return: [1, 3, 5, 78, 21, 45]

# Add element
heapq.heappush(H,8)
print(H) # return: [1, 3, 5, 78, 21, 45, 8]

# Remove element from the heap
heapq.heappop(H) 
print(H) # return: [3, 5, 78, 21, 45, 8]

# Replace an element 
heapq.heapreplace(H,6)
print(H) # return: [3, 5, 6, 78, 21, 45, 8]
```

### Sorting Algorithms

```python
#Quicksort
#select the rightmost element as the pivot.
    # Steps:
        # Step 1:  Select a pivot element
        # Step 2:  Partition the array around the pivot element. Move all the elements < pivot to the left of the pivot and move all elements >= pivot to the pivot’s right
        # Step 3:  After Step 2, the pivot element is in its correct position
        # Step 4:  Apply the quicksort recursively on the left partition
        # Step 5:  Apply the quicksort recursively on the right partition
        # Step 6:  Stop the recursion when the base case is reached. The base case is an array of size zero or one
def quickSort(array):
  if len(array)<=1:  
    return array                                                     
  else:                                                           
    pivot=array.pop()

  items_greater= []
  items_lower= []

  for item in array:
    if item > pivot:
      items_greater.append(item)
    else:
      items_lower.append(item)
  
  return quickSort(items_lower)  + [pivot]+ quickSort(items_greater)  


data = [8, 7, 2, 1, 0, 9, 6]
quickSort(data)
```

```python
# Merge Sort:
# Merge sort first divides the array into equal halves and then combines them in a sorted manner.
def merge_sort(unsorted_list):
   if len(unsorted_list) <= 1:
      return unsorted_list
# Find the middle point and devide it
   middle = len(unsorted_list) // 2
   left_list = unsorted_list[:middle]
   right_list = unsorted_list[middle:]

   left_list = merge_sort(left_list)
   right_list = merge_sort(right_list)
   return list(merge(left_list, right_list))

# Merge the sorted halves
def merge(left_half,right_half):
   res = []
   while len(left_half) != 0 and len(right_half) != 0:
      if left_half[0] < right_half[0]:
         res.append(left_half[0])
         left_half.remove(left_half[0])
      else:
         res.append(right_half[0])
         right_half.remove(right_half[0])
   if len(left_half) == 0:
      res = res + right_half
   else:
      res = res + left_half
   return res
unsorted_list = [64, 34, 25, 12, 22, 11, 90]
print(merge_sort(unsorted_list))  #RETRUN: [11, 12, 22, 25, 34, 64, 90]


#--------------------------------------------------------------------------------------------------------------
# Bubble Sort: 
#It is a comparison-based algorithm in which each pair of adjacent elements is compared
# and the elements are swapped if they are not in order.
def bubblesort(list):
   for iter_num in range(len(list)-1,0,-1):
      for idx in range(iter_num):
         if list[idx]>list[idx+1]:
            temp = list[idx]
            list[idx] = list[idx+1]
            list[idx+1] = temp
list = [19,2,31,45,6,11,121,27]
bubblesort(list)
print(list)   #retrun: [2, 6, 11, 19, 27, 31, 45, 121]





```
